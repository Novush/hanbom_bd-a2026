### **csv파일에서 여러가지 값 가져오기(Boolean List)**

````python
# Boolean List(True, False)를 이용해서 인덱싱하기 - 인덱스의 개수에 주의!
````

````python
iphone_df.loc[[True, False, True, False, True, False, True]]
````
