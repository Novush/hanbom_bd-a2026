```python
import pandas as pd

df = pd.read_csv('broadcast.csv', index_col=0)
df
```

```python
df['KBS'] > 30 # KBS 에서 시청률이 30보다 큰 값 검색 - Series
```

```python
df.loc[df['KBS'] > 30, 'KBS'] # KBS 에서 시청률이 30보다 큰 값 검색(인덱싱) - Series
```

```python
df.loc[df['KBS'] > 30, ['KBS']] # KBS 에서 시청률이 30보다 큰 값 검색(인덱싱) - DataFrame
```

```python
df[['KBS']].loc[df['KBS'] > 30] # KBS 에서 시청률이 30보다 큰 값 검색 - DataFrame (loc 즉 로케이션 명령어 없는 버전) (같은 결과지만 처리 속도가 빠름)
```
